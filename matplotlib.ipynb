```python
#      PIE PLOT()
```

```python
population=[700,560,450,510,300]
country=['india','chaina','japan','bangladesh','pakisthan']
plt.pie(population,labels=country)
plt.show()
```

```python
population=[700,560,450,510,300]
country=['india','chaina','japan','bangladesh','pakisthan']
plt.pie(population,labels=country,autopct='%1.2f%%')
plt.show()
```

```python
population=[700,560,450,510,300]
country=['india','chaina','japan','bangladesh','pakisthan']
plt.pie(population,labels=country,autopct='%1.2f%%', wedgeprops={'ec':'k'})
plt.title('country population distributation in pie plot')
plt.show()
```

```python
population=[700,560,450,510,300]
country=['india','chaina','japan','bangladesh','pakisthan']
exp =[0,0,0,0,0.4]
plt.pie(population,labels=country,autopct='%1.2f%%', wedgeprops={'ec':'k'} ,explode=exp)
plt.title('country population distributation in pie plot')
plt.show()
```

```python
population=[700,560,450,510,300]
country=['india','chaina','japan','bangladesh','pakisthan']
exp =[0,0.3,0,0,0.4]
plt.pie(population,labels=country,autopct='%1.2f%%', wedgeprops={'ec':'k'} ,explode=exp)
plt.title('country population distributation in pie plot')
plt.show()
```

```python
population=[700,560,450,510,300]
country=['india','chaina','japan','bangladesh','pakisthan']
exp =[0,0.3,0,0,0.4]
plt.pie(population,labels=country,autopct='%1.2f%%', wedgeprops={'ec':'k'} ,explode=exp,shadow=True)
plt.title('country population distributation in pie plot')
plt.show()
```
